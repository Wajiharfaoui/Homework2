```r
# Read the data
data = read.csv("./data/bankruptcy_prediction/data.csv")
head(data,20)
# Split the data 
set.seed(1)
train = sample(1:nrow(data),round(nrow(data)*0.8))
train.set = data[train,]
test.set = data[-train,]
```

```r
prop.table(table(data$Bankrupt.))
```

```r
colnames(data)
```

```r
# Neural Network model with 1 hidden layer of 30 neuros and sigmoid af 

net.1 <- neuralnet(Bankrupt.~ ROA.C..before.interest.and.depreciation.before.interest + ROA.A..before.interest.and...after.tax + Operating.Gross.Margin,
                   train.set,
                   hidden=30,  
                   rep=1,                 
                   lifesign='full',       
                   algorithm='backprop',  
                   learningrate=0.01,     
                   err.fct='ce', 
                   t='ce',          
                   act.fct="logistic",
                   stepmax=1e7,    
                   linear.output=F        
)
```

```r
# Make prediction and evaluation on train
y_test_pred <- predict(net.1, test.set[,2:96])
```

```r
# change probs to binary predictions
for (i in 1:nrow(y_test_pred)){
    if(y_test_pred[i]<0.1){
    y_test_pred[i] = 0
    } else {
    y_test_pred[i] = 1
    }
}
```

```r
# confusion matrix
table(test.set$Bankrupt., y_test_pred)
```

```r
print(paste0("accuracy: ",1319/nrow(test.set)))
```

```r
# Logistic Regression
glm.fits = glm(Bankrupt.~ROA.C..before.interest.and.depreciation.before.interest+ROA.A..before.interest.and...after.tax,train.set, family=binomial)
```
